```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(10)

datascience_jobs = {
    "Job_Title": np.random.choice(['Data Analyst', 'Jr. Data Scientist', 'Machine Learning Engineer'], size = 30),
    "Location": np.random.choice(['North America', 'Europe', 'Asian', 'Oceania'], size = 30),
    "Salary": np.random.uniform(48000, 200000, 30).round(2),
    "Average_IQ": np.random.normal(128, 25, 30).round(2)
}

df = pd.DataFrame(datascience_jobs)
df

pd.set_option('display.max_rows', 12)
```

```python
dfgroup = df.groupby("Job_Title").get_group("Jr. Data Scientist")
dfgroup
```

```python
df.groupby("Job_Title").describe()
```

```python
# In the Real World, IQ is going to Related to Salary, for Sure; How About in this Hypothetical Set?
df[['Salary', 'Average_IQ']].corr() ## Haha, tough luck!
```
